## Missing Values

```python
df = df.fillna(
    {'kuantitas':float(f'{df.kuantitas.median()}'),
     'total_harga':float(f'{df.total_harga.median()}'),
     'PPN': 0.1}

)
```

```python
df = df.drop(columns='bayar_cash')
df.isna().sum()
```

```python
df = df.astype({
    'tanggal_pembelian': 'datetime64[ns]',
    'kategori_produk': 'category',
    'kuantitas': 'int64',
    'total_harga': 'int64',
})
df.info()
```
